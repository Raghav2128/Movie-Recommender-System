```python
import numpy as np # linear algebra
import pandas as pd # data preprocessing
import ast # abstract syntax trees helps evaluate strings
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1) # returns the first row
```

```python
credits.head(1) # returns the first row
```

```python
movies.shape # returns the total rows and columns
```

```python
movies = movies.merge(credits,on='title') # merges both tables
```

```python
movies.head(1)
```

```python
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']] # creates new df with these columns
```

```python
movies.head()
```

```python
movies.isnull().sum() # checks null values for each column
```

```python
movies.dropna(inplace=True) # drops null values
```

```python
movies.duplicated().sum() # checks for any duplicates
```

```python
movies.iloc[0].genres # returns the genres column for first row
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj): # turns the string into a python list of dictionaries
        L.append(i['name']) 
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert) # calling function, apply lets you apply a function to each row/column
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj): # turns the string into a python list of dictionaries
        if counter < 3: # trying to grab first 3 actors because they are most important
            L.append(i['name']) 
            counter += 1
        else:
             break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj): # turns the string into a python list of dictionaries
        if i['job'] == 'Director':
            L.append(i['name']) 
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) # applies to each row, splits the string into a list
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x]) # Removes whitespaces between a single entry
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x)) # converts from list type to string
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower()) # converts to lower, recommended
```

```python
new_df.head()
```

```python
import nltk # natural language toolkit used for human language data
```

```python
from nltk.stem.porter import PorterStemmer # we are using it to reduce words to their root form
ps = PorterStemmer() # creates an instance of the PorterStemmer class and stores it in ps
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') # max_features is the total number of words you want, so 5000d
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() # cv.fit_transform tells CountVectorizer to learn the top 5000 words and transform each movie's tags into a numeric vector, we need to convert it to a numpy array, otherwise, we cannot understand it
```

```python
vectors # 4806 is the number of movies (vectors) and 5000 (dimensional) is the number of words we chose
```

```python
vectors[0] # returns the frequency of the top 5000 words for the first movie
```

```python
from sklearn.metrics.pairwise import cosine_similarity # to calculate similarity
```

```python
cosine_similarity(vectors).shape # calculating distance of every movie with every other movie so total number of distances calculates is 4806*4806
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] # pairs each similarity score with the index, converted to a list of tuples,gives the similar movies first based on cosine similarity,getting top 5 excluding the same movie
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] # finds the index of input movie
    distances = similarity[movie_index] # retrieves distances of the movie with every other movie
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] # explained above

    for i in movies_list:
        print(new_df.iloc[i[0]].title) # prints similar movies
```

```python
recommend('Avatar')
```
